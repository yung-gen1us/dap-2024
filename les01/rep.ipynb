**2. Как с помощью Markdown сформировать иерархический список?**

Иерархический список в Markdown создаётся с помощью символов -, * или + для ненумерованных списков и цифр с точкой для нумерованных. Уровни вложенности задаются отступами (обычно 2 или 4 пробела).

*Пример ненумерованного списка:*

- Пункт 1
  - Подпункт 1.1
  - Подпункт 1.2
- Пункт 2

*Пример нумерованного списка:*

1. Пункт 1
   1. Подпункт 1.1
   2. Подпункт 1.2
2. Пункт 2


**3. Как с помощью Markdown оформить цитату?**

Цитаты в Markdown оформляются с помощью символа > в начале строки.

*Пример:*

> Это цитата.

**5. Сколько уровней заголовка существует в Markdown?**

В Markdown существует 6 уровней заголовков, от # (самый крупный) до ###### (самый мелкий).

*Пример:*

#Заголовок 1
##Заголовок 2
###Заголовок 3
####Заголовок 4
#####Заголовок 5
######Заголовок 6

**10. Обязательно ли использовать Jupyter Notebook для анализа данных?**

Нет, не обязательно. Альтернативы:

+ IDE (PyCharm, VS Code) — для написания и отладки кода.
+ Google Colab — облачная версия Jupyter Notebook.
+ RStudio — для анализа данных на языке R.
+ Терминал — для запуска скриптов.

Jupyter Notebook удобен для интерактивной работы, визуализации и документирования, но не является единственным инструментом.
