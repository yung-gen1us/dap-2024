**Задание 3. Дан массив из 11 элементов, поменять знак у элементов, значения которых между 3 и 8**

```python
arr = np.arange(11)
print(f'Исходная матрица: {arr}')
arr[(arr > 3) & (arr < 8)] *= -1
print(f'Изменённая матрица: {arr}')
```

**Задание 4. Найти среднее заначение строки матрицы и вычесть его из каждого элемента строки**

**Задание 6. Дан массив 10x2 подеставлющий собой точки в декартовой системе координат, преобразовать в полярную**

```python
cartesian = np.random.randint(-10, 10, size=(10, 2))
print(f'Декартовы координаты:\n {cartesian}')

x = cartesian[:, 0]
y = cartesian[:, 1]
r = np.sqrt(x**2 + y**2)
theta = np.arctan2(y, x)

polar = np.column_stack((r, theta))
print(f'Полярные координаты:\n {polar}')
```

**Задание 8. Дан массив, добавить 1 к каждому элементу с индексом, заданным в другом массиве. Добавление производиться только 1 раз**

```python
size = 10
arr = np.random.randint(0, 100, size=size)
arr_index = np.random.choice(size, size=3, replace=False)
print(f'Исходный массив: {arr}')
print(f'Массив уникальных индексов: {arr_index}')
arr[arr_index] += 1
print(f'Массив после изменения: {arr}')
```

**Задание 9. Дан вектор [1, 2, 3, 4, 5], построить новый вектор с тремя нулями между каждым значением**

```python
v = np.array([1, 2, 3, 4, 5])
print(f'Исходный вектор: {v}')
zeros = np.zeros(3)
new_v = np.array([])
for i in range(len(v)):
    new_v = np.append(new_v, v[i])
    if i < len(v) - 1:
        new_v = np.append(new_v, zeros)

print(f'Новый вектор: {new_v.astype(int)}')
```

**Задание 10. Найти ближайшее к заданному значению число в заданном массиве**

```python
arr = np.arange(100)
num = np.random.uniform(0,100)
print(f'Исходное чилсо: {num}')
print(f'Исходный массив: {arr}')
closest_index = np.argmin(np.abs(arr - num))
closest_value = arr[closest_index]
print(f'Ближайшее значение в массиве: {closest_value}')
```

**2. Что будет выведено в консоль и почему?**
+ np.array([1, 4, 2, 5, 3], dtype='str')[1] + '1':
  + *Результат:* '41'
  + *Почему:* Элемент с индексом 1 это строка '4', конкатенация с '1' дает '41'
+ np.array([1, -4, 2, 0, 3], dtype='bool')[3] or np.array([1, -4, 2, 0, 3])[1]:
  + *Результат:* -4
  + *Почему:* dtype='bool' преобразует числа: 0 в False, остальные в True. Элемент с индексом 3 равен 0 → False. Операция or возвращает первый истинный операнд или последний.
+ np.array([1, -4, 2, 0, 3], dtype='bool')[3] and np.array([1, -4, 2, 0, 3])[1]:
  + *Результат:* False
  + *Почему:* Элемент с индексом 3 равен 0 → False. Операция and возвращает первый ложный операнд
+ np.array([1, -4, 2, 0, True], dtype='int64')[-1]:
  + *Результат:* 1
  + *Почему:* Так как True преобразуется в 1
+ np.array([1, -4, 2, 0, 1j], dtype='int64')[1]:
  + *Результат:* TypeError: int() argument must be a string, a bytes-like object or a real number, not 'complex'
  + *Почему:* Комплексное число 1j нельзя преобразовать в int64.
+ np.array([1, -4, 2, 0, 1j], dtype='str')[-1]:
  + *Результат:* '1j'
  + *Почему:* Комплексное число преобразуется в строку '1j'
+ np.array([1, 4, 2, 5, '3'])[-1] * 3:
  + *Результат:* '333'
  + *Почему:* Умножение строки: '3' * 3 = '333'

**10.  В чем преимущество массивов NumPy по сравнению с обычными списками. Чем обусловлено это преимущество?**
Массивы NumPy имеют несколько ключевых преимуществ перед обычными списками Python. Эти преимущества обусловлены особенностями реализации и оптимизации библиотеки NumPy.

*1. Скорость выполнения операций*
+ Преимущество: Операции с массивами NumPy выполняются значительно быстрее, чем с обычными списками Python, особенно при работе с большими объёмами данных.
+ Обоснование: NumPy написан на C и использует оптимизированные алгоритмы для работы с данными.
Операции над массивами NumPy вектор化ованы — они выполняются на уровне компилированного кода, без необходимости использовать циклы Python.
Списки Python — это динамические структуры данных, и операции с ними требуют дополнительных накладных расходов (например, проверка типов, динамическое изменение размера).


*2. Эффективное использование памяти*
+ Преимущество: Массивы NumPy занимают меньше памяти, чем списки Python, для хранения тех же данных.
+ Обоснование: NumPy хранит данные в непрерывном блоке памяти и использует фиксированные типы данных (например, int32, float64), что уменьшает накладные расходы.
Списки Python хранят ссылки на объекты, а не сами данные, что увеличивает расход памяти.


*3. Удобство работы с многомерными данными*
+ Преимущество: NumPy предоставляет удобные инструменты для работы с многомерными массивами (тензорами), что сложно или невозможно реализовать с помощью вложенных списков.
+ Обоснование: NumPy поддерживает многомерные массивы (например, матрицы, тензоры) и операции над ними (например, матричное умножение, транспонирование).
Операции над многомерными массивами выполняются эффективно и лаконично.


*4. Богатый функционал для математических операций*
+ Преимущество: NumPy предоставляет обширный набор математических функций (линейная алгебра, статистика, преобразования Фурье и др.), которые оптимизированы для работы с массивами.
+ Обоснование: Библиотека включает функции для векторных и матричных операций, что упрощает написание математического кода.
Например, умножение матриц, вычисление определителя, решение систем линейных уравнений и т. д.
